```python
# Kütüphane Yönetim Sistemi

class Library:
    def __init__(self, name, author, press_date):
        self.isim = name
        self.yazar = author
        self.basım_yılı = press_date

    def kitapListele(self):
        print(f"Kitabın ismi: {self.isim}, Yazar: {self.yazar}, Basım Yılı: {self.basım_yılı}")

    def kitapSayısı(self, kitaplar):
        print(f"Kütüphanedeki toplam kitap sayısı: {len(kitaplar)}")


# Kitap eklemek için sınıf
class AddBook(Library):
    def __init__(self, kitaplar):
        self.kitaplar = kitaplar

    def kitapEkle(self, name, author, press_date):
        yeni_kitap = Library(name, author, press_date)
        self.kitaplar.append(yeni_kitap)
        print(f"{name} kitabı kütüphaneye eklendi.")


# Ödünç verme işlemleri için sınıf
class BorrowBook:
    def __init__(self, kitaplar, odunc_kitaplar):
        self.kitaplar = kitaplar  # Kütüphanedeki kitaplar
        self.odunc_kitaplar = odunc_kitaplar  # Ödünç alınan kitaplar

    def kitapOduncVer(self, kitap_ismi):
        # Ödünç verilecek kitabı buluyoruz
        for kitap in self.kitaplar:
            if kitap.isim == kitap_ismi:
                # Kitap kütüphaneden çıkarılıyor
                self.kitaplar.remove(kitap)
                self.odunc_kitaplar.append(kitap)
                print(f"{kitap_ismi} kitabı ödünç verildi.")
                return
        print(f"{kitap_ismi} kitabı kütüphanede bulunamadı.")

    def oduncAlinanKitaplar(self):
        print("\nÖdünç Alınan Kitaplar:")
        for kitap in self.odunc_kitaplar:
            kitap.kitapListele()


# Kütüphane oluşturuluyor
kitaplar = []  # Kütüphanedeki kitaplar
odunc_kitaplar = []  # Ödünç alınan kitaplar

# Kitap ekleme
add_book = AddBook(kitaplar)
add_book.kitapEkle("Anna Karanina", "Tolstoy", 2012)
add_book.kitapEkle("Kürk Mantolu Madonna", "Sabahattin Ali", 2008)
add_book.kitapEkle("Genç Wearther'in Acıları", "Goethe", 2023)
add_book.kitapEkle("İnsan Ne İle Yaşar", "Tolstoy", 1978)
add_book.kitapEkle("Masumiyet Müzesi", "Orhan Pamuk", 2021)
add_book.kitapEkle("Kumarbaz", "Dostoyevski", 1992)

# Ödünç verme işlemi
borrow_book = BorrowBook(kitaplar, odunc_kitaplar)

# Kitap ödünç veriyoruz
borrow_book.kitapOduncVer("Anna Karanina")
borrow_book.kitapOduncVer("Kumarbaz")


# Kitapları listeleyelim
print("\nKütüphanedeki Kitaplar:")
for kitap in kitaplar:
    kitap.kitapListele()

# Ödünç alınan kitapları listeleyelim
borrow_book.oduncAlinanKitaplar()

# Kitap sayısını gösterelim
add_book.kitapSayısı(kitaplar)
```

```python
class Sehir:
    def __init__(self, isim, sicaklik):
        self.isim = isim  # Şehir adı
        self.sicaklik = sicaklik  # Şehirdeki sıcaklık

    def sicaklikGuncelle(self, yeni_sicaklik):
        """Şehre ait sıcaklık bilgisini güncelleyen metot."""
        self.sicaklik = yeni_sicaklik
        print(f"{self.isim} şehri için sıcaklık güncellendi: {self.sicaklik}°C.")

class HavaDurumu:
    def __init__(self):
        self.sehirler = []  # Şehirleri tutacak liste

    def sehirEkle(self, sehir: Sehir):
        """Yeni şehir eklemek için metot."""
        self.sehirler.append(sehir)
        print(f"{sehir.isim} şehri sisteme eklendi.")

    def sehirSorgula(self, sehir_ismi):
        """Şehir ismiyle hava durumu sorgulama."""
        for sehir in self.sehirler:
            if sehir.isim.lower() == sehir_ismi.lower():
                return sehir
        print(f"{sehir_ismi} şehri sistemde bulunamadı.")
        return None

    def tavsiyeVer(self, sicaklik):
        """Sıcaklığa göre tavsiyelerde bulunan metot."""
        if sicaklik < 0:
            return "Soğuk, sıkı giyinin."
        elif 0 <= sicaklik <= 15:
            return "Serin, mont almayı unutmayın."
        else:
            return "Hava güzel, rahat giyin."

    def havaDurumuGoster(self, sehir_ismi):
        """Belirli bir şehir için hava durumunu ve tavsiyeyi göster."""
        sehir = self.sehirSorgula(sehir_ismi)
        if sehir:
            print(f"{sehir.isim} şehri için sıcaklık: {sehir.sicaklik}°C.")
            print(self.tavsiyeVer(sehir.sicaklik))


# Uygulama başlatma
hava_durumu_sistemi = HavaDurumu()

# Kullanıcıdan şehir ve sıcaklık bilgisi alalım
while True:
    sehir_ismi = input("Şehir ismini girin (Çıkmak için 'q' tuşlayın): ")
    if sehir_ismi.lower() == 'q':
        break

    # Kullanıcıdan sıcaklık bilgisini alalım
    try:
        sicaklik = float(input(f"{sehir_ismi} şehri için sıcaklık bilgisini girin: "))
    except ValueError:
        print("Lütfen geçerli bir sayı girin.")
        continue

    # Yeni şehir ekleme
    yeni_sehir = Sehir(sehir_ismi, sicaklik)
    hava_durumu_sistemi.sehirEkle(yeni_sehir)

# Şehirlerin hava durumu ve tavsiyeleri gösterme
print("\n--- Hava Durumu Önerileri ---")
while True:
    sorgu_sehir = input("Hava durumu sorgulamak için şehir ismini girin (Çıkmak için 'q' tuşlayın): ")
    if sorgu_sehir.lower() == 'q':
        break
    hava_durumu_sistemi.havaDurumuGoster(sorgu_sehir)
```
